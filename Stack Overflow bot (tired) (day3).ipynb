# Replying

```python
import requests
import pandas as pd

url = 'https://raw.githubusercontent.com/srirammurali99/Stackoverflow-bot-on-Telegram/main/qna_telegram_bot.tsv'

df = pd.read_csv(url, sep="\t")
msg_text = []

base_url = "https://api.telegram.org/bot5160203632:AAFCzlaJanMXXg-hKlPuJU4b6ZMzupSiKB0"



def read_msg(offset):
    parameters = {
      "offset" : offset
    }
    
    resp = requests.get(base_url + "/getUpdates", data = parameters)
    data = resp.json()

    print(data)
    

    for result in data["result"]:
        msg_text.append(result["message"]["text"])
        print("The message is:", msg_text)
        send_msg(result)
    if data["result"]:
        return data["result"][-1]["update_id"] + 1

def auto_answer(message):
    answer = df.loc[df['Question'].str.lower() == message.lower()]  
    
    if(message == "/start"):
        return "Hey! This is Sriram.  Welcome to my Bot \nPlease write\
        /help to see the commands available."
    elif (message == "/help"):
        return ("You'll get the help soon\nAll the commands start with forward slash (\   /)"
    """Available Commands :-
    /youtube - To get the youtube URL
    /linkedin - To get the LinkedIn profile URL
    /gmail - To get gmail URL
    /Stack Overflow - To get Stack Overflow website""")
    elif (message == "/youtube"):
        return "Youtube Link =>\
    https://www.youtube.com/"
    elif (message == "/linkedin"):
        return "LinkedIn URL => \
        https://www.linkedin.com/in/srirammurali99"
    elif (message == "/gmail"):
        return "srirammurali99@gmail.com"
    elif (message == "/Stack Overflow" or message == "Stack Overflow"):
        return "https://stackoverflow.com/"
    elif not answer.empty:
      answer = answer.iloc[0]['Answer']
      return answer
    else:
        return "Sorry, I could not understand you !!! I am still learning and try to get better in answering."



def send_msg(message):
    text = message["message"]["text"]
    message_id = message["message"]["message_id"]
    chat_id = message["message"]["chat"]["id"]
    answer = auto_answer(text)
    
    parameters = {
      "chat_id" : chat_id,
      "text" : answer,
      "reply_to_message_id" : message_id
    }

    resp = requests.get(base_url + "/sendMessage", data = parameters)
    print(resp.json())

offset = 0

while True:
    offset = read_msg(offset)
```
